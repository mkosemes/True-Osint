```python
var_df = df.select_dtypes('number').var().idxmax()
print(f"DF1 - Plus grande variance : {var_df}")
```

```python
var_df2 = df2.select_dtypes('number').var().idxmax()
print(f"DF1 - Plus grande variance : {var_df2}")
```

#### Identification de 2 observations suspectes ou aberrantes dans chaque dataset.

```python
print("DF - 2 Observations suspectes :")
print(df.nlargest(2, var_df))
```

```python
print("DF1 - 2 Observations suspectes :")
print(df1.nlargest(2, var_df1))
```

```python
print("DF2 - 2 Observations suspectes :")
print(df2.nlargest(2, var_df2))
```
